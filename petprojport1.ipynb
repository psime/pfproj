```python

plt.figure(figsize=(6, 4))
plt.scatter(df['age'], df['fare'], alpha=0.5)
plt.axhline(df['fare'].mean(), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(df['age'].mean(), color='orange', linestyle='dashed', linewidth=2)
plt.title('Age vs Fare with Mean Lines')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.violinplot(x='pclass', y='age', data=df, inner=None, palette='pastel')
for i in df['pclass'].unique():
    median_age = df[df['pclass'] == i]['age'].median()
    plt.axhline(median_age, color='black', linestyle='dashed', linewidth=2)
plt.title('Violin Plot of Age by Passenger Class with Median Lines')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
av_age = df['age'].mean()
av_fare = df['fare'].mean()
av_fare
```

```python
# Main Arguments
plt.figure(figsize=(10, 6))
plt.scatter(
    x = df['age'],
    y = df['fare'],
    c = df['survived'],
    s = df['parch'] * 20, 
    alpha = 0.6,
    cmap = 'coolwarm',
    edgecolors = 'black',
    linewidths = 0.6)

plt.title('Titanic Dataset: Age vs. Fare\n(Size proportional to Parch, Color by Survival Status)')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.grid(True, linestyle = '--', alpha = 0.7)


# EXTRA EXTRA 
# Horizontal Line and text
plt.axvline(x = av_age, color = 'red')
plt.text(
    x = av_age + 0.2,  
    y = 450, 
    s = f'Average Age: {av_age:.1f}', 
    color = 'red',
    fontsize = 10)

# Vertical line and text
plt.axhline(y = av_fare, color = 'blue')
plt.text(
    x = 70,  
    y = av_fare + 2, 
    s = f'Average Fare: {av_fare:.1f}', 
    color = 'blue',
    fontsize = 10)


# Not extra
plt.tight_layout()
plt.show()
```
